```python
import requests
import re
import json
import pandas as pd
import numpy as np
```

```python
def getViewCnt(link):
    page = requests.get(link)
    likes = re.search("viewCount\\\\\"...(\d*).*\"author", page.text)
    try:
        likes = likes.group(1)
    except AttributeError:
        print('Can\'t get view-count , maybe the video is out!')
        return 0
    
    likes_num = int(re.sub(r'\D', "", likes))
    return likes_num
```

```python
yid = '815mgtrILMY'
getViewCnt('https://www.youtube.com/watch?v='+yid)
```

```python
moive_youtube_csv = pd.read_csv('ml-youtube.csv')

yid_list = moive_youtube_csv['youtubeId'].values.tolist()
mid_list = moive_youtube_csv['movieId'].values.tolist()
```

```python
vc_list = []
for yid in yid_list:
    vc_list.append(getViewCnt('https://www.youtube.com/watch?v='+yid))
```

```python
data = np.array([mid_list,yid_list,vc_list]).T
```

```python
df = pd.DataFrame(data=data,columns=['movieId','youtubeId','ViewCnt'])
```
